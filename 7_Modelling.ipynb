```python
X_train = df_train.drop(columns=['sentiment'])
X_test = df_test.drop(columns=['sentiment'])
y_train = df_train['sentiment']
y_test = df_test['sentiment']
```

```python
X_train = X_train[list(set(X_test.columns) & set(X_train.columns))]
for col in list(set(X_test.columns) - set(X_train.columns)):
  X_train[col] = 0
X_train = X_train[sorted(X_train.columns)]
X_test = X_test[sorted(X_test.columns)]
display(X_train)
display(X_test)
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
def classification(k_value, X_train, y_train, X_test):
  result = pd.DataFrame()
  dwknn = DWKNN(k=k_value)
  dwknn.fit(X_train, y_train)
  y_predict = dwknn.predict(X_test)
  result[f'k_{k_value}'] = pd.Series(y_predict)
  result.to_csv(f'output/7_modelling/predict_k_{k_value:02}.csv', index=False)
```
